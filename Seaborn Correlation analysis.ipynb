```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link)
df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])
```

```
# Of course, you'll add legend, readable axis ticks, and title (there is no seaborn function for that, use matplotlib syntax)
# It's up to you:


# Importer la bibliothèque matplotlib pour les graphiques
import matplotlib.dates as mdates

# Créer une figure et un objet Axes pour le graphique, avec une taille de 20x5 pouces
fig, ax = plt.subplots(figsize=(20, 5))

# Tracer un nuage de points pour les températures maximales, avec des points rouges et une étiquette "Max"
plt.scatter(x= df_weather ['DATE'], y = df_weather ["MAX_TEMPERATURE_C"], color = "red", label = "Maximun")
#idem pour Min
plt.scatter(x= df_weather ['DATE'], y = df_weather ["MIN_TEMPERATURE_C"], color = "blue", label = "Minimum ")


#obvious plt ...
plt.title("Températures")
plt.xlabel("DATE")
plt.ylabel("Températures")
plt.legend()

# Définir le localisateur pour les dates sur l'axe des x, pour afficher un tick par mois
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# Définir le format des dates pour les ticks sur l'axe des x, pour afficher le mois et l'année
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))

# Limite l'axe x à la période de janvier 2019 à décembre 2019
plt.xlim([df_weather['DATE'].min(), df_weather['DATE'].max()])

# Pour éviter les espaces vides avant les dates
plt.tight_layout()

plt.show()
```

## Pair Plot - Colorize the opinions
Another way, and more visual way, to see correlation, is just to plot each column 2 by 2, that's a pairplot. But it's slow to compute on the whole dataset.

- Creates a slice of 10 columns (including the "OPINION" column)
- Creates a sample of this slice
- Draw a Pairplot on this sample, and colorize it by `OPINION`

```
# It's your turn:
colonnes = ["DATE","MAX_TEMPERATURE_C", "MIN_TEMPERATURE_C","TEMPERATURE_MORNING_C","TEMPERATURE_NOON_C","TEMPERATURE_EVENING_C","HEATINDEX_MAX_C","DEWPOINT_MAX_C","WINDTEMP_MAX_C","OPINION"]
df_w_col = df_weather[colonnes]

df_w_test = df_w_col.sample(40)

sns.pairplot(data=df_w_test,
             hue="OPINION")
plt.show()
```

# More and more...
Draw 4 different charts of your choice on the same dashboard. You're an analyst, you have to be force of proposal. Your goal here is to represent which season has the best weather.

Remember, a MatPlotLib "figure" is like a dashboard including subplots. So you have to create a MatPlotLib figure with subplots. And these subplots will be Seaborn's charts.

```
# Your code here

Saisonnalité = [df_weather['DATE'].dt.month.isin([12,1,2]),
           df_weather['DATE'].dt.month.isin([3,4,5]),
           df_weather['DATE'].dt.month.isin([6,7,8]),
           df_weather['DATE'].dt.month.isin([9,10,11])]
nomValeur = ['Hiver','Printemps','Eté','Automne']

df_weather['SAISON'] = np.select(Saisonnalité,nomValeur)
df_weather.head()
```

```
fig, ax = plt.subplots(2,2,figsize = (20,8))
fig.suptitle("SAISONNALITÉ OPTI")

sns.boxplot(data=df_weather,x='SAISON',y="MAX_TEMPERATURE_C",ax=ax[0, 0])

ax[0,0].set_title("Températures MAX/SAISON")
ax[0,0].set_xlabel("Saisons")
ax[0,0].set_ylabel("Températures")

sns.boxplot(x='SAISON',y='HUMIDITY_MAX_PERCENT', data = df_weather,ax=ax[0, 1])
ax[0,1].set_title("Humidité max/Saison")
ax[0,1].set_xlabel("Saisons")
ax[0,1].set_ylabel("densité nuageuse ")

sns.barplot(x='SAISON',y='HEATINDEX_MAX_C',data = df_weather,ax=ax[1, 0])
ax[1,0].set_title("Indice UV/Saison")
ax[1,0].set_xlabel("Saisons")
ax[1,0].set_ylabel("Indice UV")

sns.barplot(data=df_weather,x='SAISON',y='CLOUDCOVER_AVG_PERCENT',ax=ax[1, 1])
ax[1, 1].set_title("temps nuageux/Saison")
ax[1, 1].set_xlabel("Saisons")
ax[1, 1].set_ylabel("Couverture Nuageuse en prct")


plt.show()

```
